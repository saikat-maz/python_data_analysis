```python
 import numpy as np
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sb
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
zx=sb.countplot(x='Gender',data=df)
for bars in zx.containers:
    zx.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen
```

```python
sb.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
# From above graphs we can see that most of the buyers are females and even the purchasing power of females are greaterthan men
```

```python
ax=sb.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
# From above graph we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sb.set(rc={'figure.figsize':(16,6)})
sb.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(16,6)})
sb.barplot(x='State',y='Amount',data=sales_state)
```

```python
# From above graph we can see that most of the orders and total sales amount are from uttar pradesh,maharashtra and karnataka
```

```python
nx=sb.countplot(x='Marital_Status',data=df)
for bars in nx.containers:
    nx.bar_label(bars)
sb.set(rc={'figure.figsize':(4,4)})
```

```python
sales_ms=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.set(rc={'figure.figsize':(6,4)})
sb.barplot(x='Marital_Status',y='Amount',data=sales_ms,hue='Gender')
```

```python
# From above graph we can see that most of the buyers are married(women) and they have high purchasing power
```

```python
sb.set(rc={'figure.figsize':(20,5)})
ax=sb.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(x='Occupation',y='Amount',data=sales_occ)
```

```python
# From above graph we can see that most of the buyers are working in IT,Healthcare and Aviation sectors.
```

```python
sb.set(rc={'figure.figsize':(26,7)})
ax=sb.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.set(rc={'figure.figsize':(35,10)})
sb.barplot(x='Product_Category',y='Amount',data=sales_prod)
```

```python
# From above graph we can see that most of the sold products are from food,clouthing and electronics category.
```
